```python
import glob
import xarray as xr
from mom6 import DATA_PATH
```

```python
file_list = glob.glob(f'{DATA_PATH}hindcast/*.nc')
```

```python
for file in file_list :
    ds = xr.open_dataset(file)
    coords = list(ds.coords)
    vars = [var for var in list(ds.variables) if var not in coords]

    for i in range(len(ds.init)):
        year = ds.isel(init=i)['init.year'].data
        month = ds.isel(init=i)['init.month'].data
        ds.isel(init=i).to_netcdf(f'{DATA_PATH}forecast/{vars[0]}_forecast_i{year:04d}{month:02d}.nc')

```
